```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import os
import matplotlib.image as mpimg
plt.ioff()
import pandas as pd

info = pd.read_csv("patient_info.csv")
info = info.loc[info["yeni_koordinat"] != "Sil"]
image_dir = os.path.join("data", "tomo", "images")
crop_dir = os.path.join("data", "tomo", "cropped_images")

if not os.path.exists(crop_dir):
    os.makedirs(crop_dir)
    os.makedirs(os.path.join(crop_dir,"1_root_images"))
    os.makedirs(os.path.join(crop_dir,"2_root_images"))

one_root_dir = os.path.join(crop_dir,"1_root_images")
two_root_dir = os.path.join(crop_dir,"2_root_images")
```

```python
for patient in info["patient_id"]:
    patient_dir = os.path.join(image_dir, patient)
    check_root_num_15 = (info[info["patient_id"] == patient]["15_root_num"]==1).values[0]
    image_range_15 = (info[info["patient_id"] == patient]["15_exact_img"]).values[0]
    crop_coord_15 = (info[info["patient_id"] == patient]["15_crop_coordinate"]).values[0]

    image_start_15, image_end_15 = image_range_15.split("-")
    image_start_15, image_end_15 = int(image_start_15), int(image_end_15)
    step_size_15 = (image_end_15-image_start_15)//10
    image_end_15 = image_start_15 + step_size_15*10

    coord_y_15, coord_x_15 = crop_coord_15.split("-")
    coord_y_15, coord_x_15 = int(coord_y_15), int(coord_x_15)
    y_len, x_len = 20, 30

    print(patient, "15:", step_size_15, image_start_15, image_end_15)

    for i in range(image_start_15,image_end_15,step_size_15):

        curr_image = os.path.join(patient_dir, f"{i}.jpeg")
        img = mpimg.imread(curr_image)
        fig, ax = plt.subplots()
        plt.axis('off')

        img_15_plot = ax.imshow(img[coord_y_15:(coord_y_15+y_len), (coord_x_15):(coord_x_15+x_len)],
                                cmap=plt.cm.bone)
        if check_root_num_15:
            root_dir = one_root_dir
        else:
            root_dir = two_root_dir
        plt.savefig(os.path.join(root_dir, f"{patient}_{i}_15.jpeg"), bbox_inches='tight', pad_inches=0)
        plt.close(fig)

    check_root_num_25 = (info[info["patient_id"] == patient]["25_root_num"]==1).values[0]
    image_range_25 = (info[info["patient_id"] == patient]["25_exact_img"]).values[0]
    crop_coord_25 = (info[info["patient_id"] == patient]["25_crop_coordinate"]).values[0]

    image_start_25, image_end_25 = image_range_25.split("-")
    image_start_25, image_end_25 = int(image_start_25), int(image_end_25)
    step_size_25 = (image_end_25-image_start_25)//10
    image_end_25 = image_start_25 + step_size_25*10

    coord_y_25, coord_x_25 = crop_coord_25.split("-")
    coord_y_25, coord_x_25 = int(coord_y_25), int(coord_x_25)
    y_len, x_len = 20, 30

    print(patient, "25:", step_size_25, image_start_25, image_end_25)

    for i in range(image_start_25,image_end_25,step_size_25):

        curr_image = os.path.join(patient_dir, f"{i}.jpeg")
        img = mpimg.imread(curr_image)
        fig, ax = plt.subplots()
        plt.axis('off')
        
        img_25_plot = ax.imshow(img[coord_y_25:(coord_y_25+y_len), (coord_x_25):(coord_x_25+x_len)],
                                cmap=plt.cm.bone)
        if check_root_num_25:
            root_dir = one_root_dir
        else:
            root_dir = two_root_dir
        plt.savefig(os.path.join(root_dir, f"{patient}_{i}_25.jpeg"), bbox_inches='tight', pad_inches=0)
        plt.close(fig)

```
