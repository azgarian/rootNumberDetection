```python
patient_rerun2 = patient_rerun[12:]
```

```python
import matplotlib
matplotlib.use('Agg')

import matplotlib.pyplot as plt
import pydicom

image_dir = os.path.join("/Users", "ecemazgari", "Documents", "aiproject")
dicom_dir = os.path.join("/Volumes", "Cem's HD", "Ecem Tez Tomografi Hastaları")
images_path = os.path.join(image_dir, "images")

for patient in patient_rerun:

    patient_dir = os.path.join(images_path, patient)
        
    dicom_file = os.path.join(dicom_dir, f"{patient}.dcm")
    curr_dicom = pydicom.dcmread(dicom_file)

    upper_limit, lower_limit = 300, 200  

    fig_interval = 1
    num = int((upper_limit - lower_limit) / fig_interval)

    print(patient_dir, flush=True)

    for i in range(num):
        plt.figure()
        plt.axis('off')
        img=plt.imshow(curr_dicom.pixel_array[lower_limit+i*fig_interval], cmap=plt.cm.bone)
        plt.savefig(os.path.join(patient_dir, f"{lower_limit+i*fig_interval}.jpeg"), bbox_inches='tight', pad_inches=0)
        plt.close()
```

```python
for patient in patient_rerun_koron2:
    dicom_file = f"{dicom_dir}/{patient}.dcm"
    image_folder = f"{image_dir}/{patient}"
        
    curr_dicom = pydicom.dcmread(dicom_file)

    upper_limit, lower_limit = 300, 0  

    fig_interval = 1
    num = int((upper_limit - lower_limit) / fig_interval)

    print(image_folder, flush=True)

    for i in range(num):
        plt.figure()
        plt.imshow(np.transpose(curr_dicom.pixel_array, (1,0,2))[lower_limit+i*fig_interval], cmap=plt.cm.bone)
        plt.savefig(f"{image_folder}/{lower_limit+i*fig_interval}.png")
        plt.close()
```
