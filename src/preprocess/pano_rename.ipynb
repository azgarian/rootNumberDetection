```python
import os
import shutil
import pandas as pd

# get the names of the jpg files in the pano_raw directory
pano_dir = os.path.join("data", "pano")
raw_dir = os.path.join(pano_dir, "pano_raw")
dest_dir = os.path.join(pano_dir, "images")
os.makedirs(dest_dir, exist_ok=True)
jpg_files = [file for file in os.listdir(raw_dir) if file.endswith(".jpg")]

# using patient_to_id.csv, get the patient id for each patient name
csv_file = os.path.join("data", "patient_to_id.csv")
df = pd.read_csv(csv_file)

# Create a dictionary to map patient names to patient IDs
patient_dict = dict(zip(df['file_name'], df['patient_id']))

# Iterate over the jpg_files list and get the patient ID for each patient name
patient_names = [patient_name.lower().replace("_pano.jpg", "") for patient_name in jpg_files]

for patient_name in patient_names:
    if patient_name in patient_dict:

        # Copy the JPG file to the images directory with the patient ID as the filename
        source_path = os.path.join(raw_dir, (patient_name + "_pano.jpg"))
        destination_path = os.path.join(dest_dir, str(patient_dict[patient_name] + ".jpg"))
        shutil.copy(source_path, destination_path)

```

```python
for name in patient_dict.keys():
    if name not in patient_names:
        print(name)
```
